## Verificando a existência de duplicadas e dados nulos e excluindo-os

```python
# verificar dados duplicados

emprestimos.duplicated().sum()
```

```python
# excluir dados duplicados

emprestimos.drop_duplicates(inplace=True)
```

## Importando dados dos exemplares do acervo

```python
# Importei baixando o arquivo e upando no colab em vez de direto do github, pq deste modo estava dando um erro e eu estava com dificuldades de solucionar
exemplares = pd.read_parquet('dados_exemplares.parquet')
exemplares
```

## Mesclando os datasets

```python
dados_biblioteca = pd.merge(emprestimos, exemplares, on='codigo_barras')
dados_biblioteca
```

### Usando condições para definir as áreas de conhecimento dos livros com base no número da localização

000 a 099: Generalidades. Ciência e conhecimento.

---


100 a 199: Filosofia e psicologia.

---


200 a 299: Religião.

---


300 a 399: Ciências sociais.

---


400 a 499: Classe vaga. Provisoriamente não ocupada.

---


500 a 599: Matemática e ciências naturais.

---


600 a 699: Ciências aplicadas.

---


700 a 799: Belas artes.

---


800 a 899: Linguagem. Língua. Linguística.

---


900 a 999: Geografia. Biografia. História.

```python
def localizar_area(localizacao):
  if localizacao >= 000 and localizacao < 99:
    return 'Generalidades. Ciência e conhecimento'
  elif localizacao >= 100 and localizacao < 199:
    return 'Filosofia e psicologia'
  elif localizacao >= 200 and localizacao < 299:
    return 'Religião'
  elif localizacao >= 300 and localizacao < 399:
    return 'Ciências Sociais'
  elif localizacao >= 400 and localizacao < 499:
    return 'Classe vaga. Provisoriamente não ocupada'
  elif localizacao >= 500 and localizacao < 599:
    return 'Matemática e ciências naturais'
  elif localizacao >= 600 and localizacao < 699:
    return 'Ciências aplicadas'
  elif localizacao >= 700 and localizacao < 799:
    return 'Belas artes'
  elif localizacao >= 800 and localizacao < 899:
    return 'Linguagem. Língua. Linguística'
  elif localizacao >= 900 and localizacao < 999:
    return 'Geografia. Biografia. História'
  else:
    return 'Não identificado'
```

```python
dados_biblioteca['area_conhecimento'] = dados_biblioteca['localizacao'].apply(localizar_area)
dados_biblioteca.head()
```

```python
import matplotlib.pyplot as plt

# Extraindo o ano
dados_biblioteca['ano'] = dados_biblioteca['data_emprestimo'].dt.year

# Contando a quantidade de empréstimos por ano
emprestimos_por_ano = dados_biblioteca.groupby('ano').size()

# Criando o gráfico de linhas
plt.figure(figsize=(10, 6))
plt.plot(emprestimos_por_ano.index, emprestimos_por_ano.values, marker='o', linestyle='-', color='b', label='Empréstimos de Livros')

# Adicionando título e rótulos aos eixos
plt.title('Quantidade de Empréstimos de Livros por Ano')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Empréstimos')
plt.legend()

# Exibindo o gráfico
plt.grid(True)
plt.show()
```

Como pode ser observado no gráfico, a maior quantidade de emprestimos foi realizada no ano de 2014, com quase 225.000 empréstimos realizados naquele ano. Em contrapartida, em 2020 houve a menor quantidade de empréstimos registrada, apenas 25.000, esse fato pode ser devido à pandemia de Covid-19 que ocorreu em 2020 e causou o afastamento dos alunos, não só das universidades, das salas de aula, causando, assim, essa drástica queda nos empréstimos de livros, quando comparados aos anos anteriores.

```python
# Extraindo o ano
dados_biblioteca['mes'] = dados_biblioteca['data_emprestimo'].dt.month

# Contando a quantidade de empréstimos por mês
emprestimos_por_mes = dados_biblioteca.groupby('mes').size()

# Criando o gráfico de linhas
plt.figure(figsize=(10, 6))
plt.plot(emprestimos_por_mes.index, emprestimos_por_mes.values, marker='o', linestyle='-', color='b', label='Empréstimos de Livros')

# Adicionando título e rótulos aos eixos
plt.title('Quantidade de Empréstimos de Livros por Mês')
plt.xlabel('Mês')
plt.ylabel('Quantidade de Empréstimos')
plt.legend()

# Exibindo o gráfico
plt.grid(True)
plt.show()
```

Como é possível observar, a maior quantidade de empréstimos é realizada no mês de agosto, quando foram realizados mais de 225.000 empréstimos, e a menor quantidade de empréstimos foi realizada nos meses de Janeiro e dezembro, com pouco mais de 50.000 e 75.000, respectivamente. Isto pode ser devido ao fato de esses meses serem, em geral, quando ocorrem as férias de fim de ano, bem como os meses de Junho e Julho, que marcam o recesso do meio do ano.

Considerando que os meses de Janeiro, Junho, Julho e Dezembro se mostraram serem os meses de menor demanda, é possível gerenciar a programação de férias dos colaboradores para que ocorram nesses meses, bem como programar atividades que não sejam de atendimento ao usuário para estes períodos.

```python
# Extraindo o ano
dados_biblioteca['horario'] = dados_biblioteca['data_emprestimo'].dt.hour

# Contando a quantidade de empréstimos por mês
emprestimos_por_hora = dados_biblioteca.groupby('horario').size()

# Criando o gráfico de linhas
plt.figure(figsize=(10, 6))
plt.bar(emprestimos_por_hora.index, emprestimos_por_hora.values, color='b', label='Empréstimos de Livros')

# Adicionando título e rótulos aos eixos
plt.title('Quantidade de Empréstimos de Livros por Hora')
plt.xlabel('Horário')
plt.ylabel('Quantidade de Empréstimos')
plt.legend()

# Exibindo o gráfico
plt.grid(True)
plt.show()
```

Os horários com maior quantidade de empréstimos ao longo do dia são por volta das 10, 16, 17 e 18 horas.
Por outro lado, os horários em qua há menos empréstimos realizados são por volta das 7 e das 22 horas, horários em que poderiam ser alocadas as demais atividades que não sejam de atendimento ao usuário. No entanto, por marcarem o início e final de expediente, uma alternativa seria por volta das 8 e das 19 horas, onde há um fluxo mais moderado de empréstimos.
